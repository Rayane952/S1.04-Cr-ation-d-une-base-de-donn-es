Nous avons créer 3 tables.
Dans la premiére Table que nous avons nommé "Passagers":

Nous avons 5 attributs.

'Name' : avec le type varchar(chaine de caracteres) qui est elle une clé primaire car elle n'a ni emplacement vide (not null) ni de tuples repetitives (c'est a dire qu'il ya que des tuples uniques).Cet attribut liste tous les prenoms des passagers.

'Sex' : avec le type varchar(10), cet attribut correspond au sexe des passagers si il est 'male' ou 'female'.

'Age' :  il est de type Float(un nombre a virgule), c'est l'age des passagers.

'SibSp' :  de type integer (valeur entiere) cet attribut indique le nombre d'epoux/femmes parmi les passagers. 

'Parch' :  egalement de type interger et indique le nombre de parents ou enfants passagers.

Dans la dexieme table que nous avons nommé "Trajet":

Nous avons aussi 5 attributs.

'PassengerID' :  de type interger cet attribut est aussi la clé primaire (elle a aucun emplacement vide et tous les tuples sont uniques), cet attribut associe une valeur entiere propre à chaque passager

'Pclass' :  de type int(variant de 1 à 3) qui quant à lui indique le niveau de classe du passager en fonction de leurs moyens.

'Ticket' : de type varchar qui correspond tout simplement au numéro du passager.

'Fare' : de type float(nombre à virgules), Cet attribut indique le prix payé par un passager pour effectuer la croisière titanic.

'Cabin' : de type varchar chaine de caractère ou str) liste la/les cabine(s) associés au passagers. 

Pour finir, la dernière que nous avons nommé “titanic" est composée de 4 attributs:

'Name' : de type varchar(str) qui dans ce tableau ci est une clé étrangère(une clé étrangère est une clé primaire dans une autre relation/table)

'Passengerld' : de type integer qui est lui aussi une clé étrangère (car déjà primaire dans la relation “Trajet").

'Survived' : de type integer(variants de 0 à 1).Si le chiffre correspond à 1, le passager a survécu et si le chiffre correspond à 0 alors il n’a pas survécu.

'embarked' : de type varchar(str) qui attribue des lettres aux embarcations.


(NB:Les clés primaires et étrangères sont très utiles pour faire des jointures entre les différentes relations/tables)

Voici ci-dessous Notre Script : creé 


```python
%%sql 
drop table if exists titanic;
drop table if exists passagers;
drop table if exists trajet;

create table Trajet(                    
        PassengerId int not null,
        Pclass int,
        Ticket varchar(50),
        Fare float,
        Cabin varchar(20),
        primary key(PassengerId));

create table Passagers(  
        PassengerId int not null,
        name varchar(100) not null,
        Sex varchar(10),
        Age float,
        SibSp int,
        Parch int,
        primary key(Name),
        foreign key(PassengerId) references Trajet(PassengerId));

create table titanic(       
        PassengerId int not null,
        Name varchar(100) not null,
        Survived char(4),
        Embarked varchar(1),
        foreign key(Name) references Passagers(Name),
        foreign key(PassengerId) references Trajet(PassengerId));
```

```python
def parse_line(ligne, sep, num_colonne, ty):  # Programme du Cours 7 'Tableaux' 'Lecture et écriture formatée'
    li = ligne.split(sep)                     # Pour lire le contenu du Fichier 'titanic_train.csv'
    if ty == "int":
        return (int(li[num_colonne]))
    elif ty == "float":
        return (float(li[num_colonne]))
    elif ty == "str":
        return (li[num_colonne])
    elif ty == "bool":
        return (bool(li[num_colonne]))
    else:
        print("Erreur type non reconnu.")


f_input = open( 'titanic_train.csv' , 'r') # On Ouvre le fichier en 'Read' pour lire les données et les 
                                           # mémoriser dans les variables qui represente les attributs du fichier csv.
PassengerID = []
Survived = []
Pclass = []
name1 = []
name2 = []
name3 = []
Sex = []
Age = []
SibSp = []
Parch = []
Ticket = []
Fare = []
Cabin = []
Embarked = []
nameF = []
f_input.readline()
li = f_input.readline()
while( '' != li ) :        # ecrit les données memorisé dans les tableau des variables
    PassengerID.append(parse_line(li, ',', 0, "int"))
    Survived.append(parse_line(li, ',', 1, "str"))
    Pclass.append(parse_line(li, ',', 2, "int"))
    name1.append(parse_line(li, ',', 3, "str"))
    name2.append(parse_line(li, ',', 4, "str"))
    Sex.append(parse_line(li, ',', 5, "str"))
    Age.append(parse_line(li, ',', 6, "str"))
    SibSp.append(parse_line(li, ',', 7, "int"))
    Parch.append(parse_line(li, ',', 8, "int"))
    Ticket.append(parse_line(li, ',', 9, "str"))
    Fare.append(parse_line(li, ',', 10, "float"))
    Cabin.append(parse_line(li, ',', 11, "str"))
    Embarked.append(parse_line(li, ',', 12, "str"))
    li = f_input.readline() 
f_input.close() # on ferme le fichier 'titanic_train.csv'

i = 0                       # Impossible d'afficher le 'name' complet 
while i < len(name1) :      # j'ai du faire un mini programme pour faire en sorte que le name apparaisse completement
    name3.append(name1[i] + name2[i]) # 
    i = i + 1


f_output = open('TRAJET.csv', 'w') # Ouverture du fichier TRAJET.csv ou Creation de celui-ci si il est inexistant
i = 0
while i < len(PassengerID):
    f_output.write (str(PassengerID[i]) + ',' + str(Pclass[i]) + ',' + Ticket[i] + ',' + str(Fare[i]) + ',' + Cabin[i] + '\n')
    i = i + 1
f_output.close()

f_output = open('PASSAGERS.csv', 'w') # Ouverture du fichier PASSAGERS.csv ou Creation de celui-ci si il est inexistant
i = 0
while i < len(PassengerID) :
    f_output.write (str(PassengerID[i]) + ',' + name3[i] + ',' + Sex[i] + ',' + Age[i] + ',' + str(SibSp[i]) + ',' + str(Parch[i]) + '\n')
    i = i + 1
f_output.close()


f_output = open('TITANIC.csv', 'w') # Ouverture du fichier TITANIC.csv ou Creation de celui-ci si il est inexistant
i = 0
while i < len(PassengerID):
    f_output.write (str(PassengerID[i]) + ',' + str(name3[i]) + ',' + (Survived[i]) + ',' + str(Embarked[i]))
    i = i + 1
f_output.close()
```
